### Seasonal Decomposition 

##### Is there Seasonality in the Time Series? What is the best m?

- The seasonal period/order (m) is likely the lag position where autocorrelation or partial autocorrelation sharply decreases, signaling the presence of a seasonal pattern in the data.

- If the ACF and PACF suggest different values for m, it could imply that the seasonality in the time series is not straightforward or does not strictly follow a single pattern.

- If one plot indicates a value for m, while the other does not, it may suggest the presence of more complex temporal patterns, possibly involving multiple seasonal components or a combination of trends.

```python
def find_best_m(series):
     
    # Plot ACF using the plot_acf function on undifferenced y_train
    plot_acf(series, lags=int(0.1 * len(series)))
    plt.title('ACF Plot of y_train for determining "m"')
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.show()
    
    # Plot PACF using the plot_acf function on undifferenced y_train
    plot_pacf(series, lags=int(0.1 * len(series)))
    plt.title('PACF Plot of y_train for determining "m"')
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.show()
    

# Call the function with your y_train data
find_best_m(df[target])
```

##### Auto Seasonal Decomposition

```python
# Function to automate decomposition model and seasonality period selection
def auto_decompose(series):
    # Perform ADF test for stationarity
    result_adf = adfuller(series)
    is_stationary = result_adf[1] < 0.05  # Check if p-value is less than 0.05

    # Choose decomposition model based on stationarity
    selected_model = 'multiplicative' if is_stationary else 'additive'

    # retrieve selected period
    selected_period = m
    
    print(f"Selected Model: {selected_model}")
    print(f"Selected Period: {selected_period}")

    return selected_model , selected_period

# Perform automated decomposition model and seasonality period selection
selected_model, selected_period = auto_decompose(df[target])

# Perform seasonal decomposition with the selected model and period
result = seasonal_decompose(df[target], model=selected_model, period=selected_period)

# Plot the decomposed components
result.plot()
plt.show()
```

### Time Series Modeling with Statistical Models (with External Variables)

##### What is the best d?

The function determines the optimal differencing parameter ('d') using the ADF test, aiming for the first order 'd' with p-value below 0.05. 

```python
def find_best_d(y_train):
    
    # Initialize variables to store the best d value and corresponding p-value
    best_d = None
    best_p_value = 1.0  # Initialize with a value higher than 0.05

    # Iterate over different values of d
    for d in range(1, 4):  # Adjust the range based on your data and requirements
        # Create a differenced series with the current d value
        differenced_series = y_train.diff(d).dropna()

        # Perform Augmented Dickey-Fuller test
        adf_result = adfuller(differenced_series)

        # Extract the p-value from the ADF test result
        p_value = adf_result[1]

        # Update the best d value if the current p-value is lower
        if p_value < best_p_value:
            best_p_value = p_value
            best_d = d

            # Break out of the loop if the first value lower than 0.05 is found
            if best_p_value < 0.05:
                
                # Plot the differenced data for visualization
                plt.figure(figsize=(11, 2))
                plt.plot(y_train.diff(best_d).dropna(), label=f'Differenced Series (d={best_d})')
                plt.title('Differenced Data for ADF Test')
                plt.xlabel('Time')
                plt.ylabel('Differenced Values')
                plt.legend()
                plt.show()
                
                break

    # Print the best d value and its corresponding ADF p-value
    print(f'Best d value: {best_d} (ADF p-value: {best_p_value})')

    return best_d

# Call the function with your y_train data
best_d = find_best_d(y_train)
```

```python
# Set value
d = best_d

print(f"Estimated d: {d}")
```

##### Is the time series a random walk?

- In a random walk, values are usually interdependent with a persistence that leads to a slower autocorrelation decay. Abrupt decay, particularly after the 0th lag, signals a lack of this persistence and indicates that the series may not demonstrate the typical characteristics of a random walk.

- While traditional statistical models may face challenges, various forecasting methods, including both statistical models like ARIMA and machine learning or deep learning approaches, can be effective in capturing and predicting patterns in such time series data.

```python
def check_if_random_walk(y_train):
    
    # Plot ACF using the plot_acf function on undifferenced y_train
    plot_acf(y_train, lags=int(0.1 * len(y_train)))
    plt.title('ACF Plot of y_train for determining if Time Series is a Random Walk')
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.show()
    
# Call the function with your y_train data
check_if_random_walk(y_train)
```

```python
# Set Boolean value
random_walk = True

print(f"Is Time Series likely a random walk?: {random_walk}")
```

##### Set m to determine best D!

The function determines the optimal differencing parameter ('D') using the ADF test, aiming for the first order 'D' with p-value below 0.05. 

```python
def find_best_D(y_train, m):
    # m is the seasonal periodicity (number of observations per seasonal cycle)
    
    # if seasonal periodicity does not exist then D is 0
    if m == 0:
        best_D = 0
        print("Seasonal Differencing order does not exist.")
        print(f'Best D value: {best_D}')
        return best_D

    # Initialize variables to store the best D value and corresponding p-value
    best_D = None
    best_p_value = 1.0  # Initialize with a value higher than 0.05

    # Iterate over different values of D
    for D in range(1, 4):  # Adjust the range based on your data and requirements
        # Create a seasonally differenced series with the current D value
        seasonally_differenced_series = y_train.diff(m * D).dropna()

        # Perform Augmented Dickey-Fuller test
        adf_result = adfuller(seasonally_differenced_series)

        # Extract the p-value from the ADF test result
        p_value = adf_result[1]

        # Update the best D value if the current p-value is lower
        if p_value < best_p_value:
            best_p_value = p_value
            best_D = D

            # Break out of the loop if the first value lower than 0.05 is found
            if best_p_value < 0.05:
                
                # Plot the differenced data for visualization
                plt.figure(figsize=(11, 2))
                plt.plot(y_train.diff(best_D).dropna(), label=f'Differenced Series (D={best_D})')
                plt.title('Differenced Data for ADF Test')
                plt.xlabel('Time')
                plt.ylabel('Differenced Values')
                plt.legend()
                plt.show()
                
                break

    # Print the best D value and its corresponding ADF p-value
    print(f'Best D value: {best_D} (ADF p-value: {best_p_value})')
    
    return best_D

# Call the function with your y_train data and best_m
best_D = find_best_D(y_train, m)
```

```python
# Set value
D = best_D

print(f"Estimated D: {D}")
```

##### Is this a MA(q) process? What is the best q?

Abrupt decay after the 0th lag in the ACF plot typically suggests that the time series may not follow a Moving Average (MA(q)) process. Conversely, observing a gradual decay until the nth lag followed by abrupt decay can indicate the potential order of the MA process, suggesting that 'q' might be the lag at which the decay becomes abrupt. This pattern is commonly employed for identifying the order of the moving average component in a time series.

```python
def find_best_q(y_train_diff):
    
    # Plot ACF using the plot_acf function for differenced series
    plot_acf(y_train_diff, lags=int(0.1 * len(y_train_diff)))
    plt.title('ACF Plot of Differenced y_train for determining "q"')
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.show()

# Call the function with your differenced y_train data
best_q = find_best_q(y_train.diff(d).dropna())
```

##### SARIMAX

- The code defines functions to optimize parameters, fit, and evaluate a SARIMAX model with exogenous variables. Additionally, it creates a leaderboard of SARIMAX model results. 
- SARIMAX requires future values of exogenous variables for accurate forecasting. Recursive forecasting is possible if a DataFrame containing predicted future values for required exogenous variables is available, allowing SARIMAX to make predictions based on the forecasted exogenous variable values.

##### Time Series Modeling using parameters selected via Manual Inspection

```python
# Function to evaluate model performance
def evaluate_model(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    return mse, rmse, mae

# Function to fit SARIMAX model with exogenous variables
def fit_sarimax_model(y_train, exog_train, order, seasonal_order):
    sarimax_model = SARIMAX(y_train, exog=exog_train, order=order, seasonal_order=seasonal_order, freq=freq, enforce_stationarity=False)
    sarimax_result = sarimax_model.fit(disp=False, method='lbfgs', maxiter=100)  
    return sarimax_result

# Function to visualize predictions
def visualize_predictions(y_train, pred_train, y_val, pred_val, model_name):
    plt.figure(figsize=(11, 2))

    # Plot actual, training set predictions, validation set predictions
    plt.plot(y_train.index, y_train, label='Actual', color='blue')
    plt.plot(y_train.index, pred_train, label='Training Set Prediction', color='red', linestyle='--')
    plt.plot(y_val.index, y_val, label='Actual (Validation)', color='green')
    plt.plot(y_val.index, pred_val, label='Validation Set Prediction', color='orange', linestyle='--')

    plt.title(f'{model_name} - Actual vs Predicted')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()

    plt.show()

# Function to generate SARIMAX leaderboard with exogenous variables
def generate_sarimax_leaderboard(y_train, exog_train, y_val, exog_val, p, d, q, P, D, Q, m):
    # Create a list to store results
    results_list = []

    # Define different SARIMA configurations
    if m == 1:
        orders = [
            {'name': 'SARIMAX', 'order': (p, d, q), 'seasonal_order': (0, 0, 0, 0)}
        ]
    else:
        orders = [
            {'name': 'SARIMAX', 'order': (p, d, q), 'seasonal_order': (P, D, Q, m)}
        ]

    # Iterate through different configurations
    for model_params in orders:
        order = model_params['order']
        seasonal_order = model_params['seasonal_order']

        # Fit SARIMAX model with exogenous variables
        sarimax_model = fit_sarimax_model(y_train, exog_train, order, seasonal_order)

        # Generate predictions for training and validation sets
        y_train_pred  = sarimax_model.get_prediction(start=0, end=len(y_train)-1, exog=exog_train).predicted_mean
        y_val_pred = sarimax_model.get_prediction(start=len(y_train), end=len(y_train)+len(y_val)-1, exog=exog_val).predicted_mean

        # Evaluate model performance
        mse_train, rmse_train, mae_train = evaluate_model(y_train, y_train_pred)
        mse_val, rmse_val, mae_val = evaluate_model(y_val, y_val_pred)

        # Store results in the list
        results_list.append({
            'Model': model_params['name'],
            'Params': f'Order={order}, Seasonal_Order={seasonal_order}',
            'AIC': sarimax_model.aic,
            'MSE_Train': mse_train,
            'RMSE_Train': rmse_train,
            'MAE_Train': mae_train,
            'MSE_Val': mse_val,
            'RMSE_Val': rmse_val,
            'MAE_Val': mae_val
        })
        
        # Visualize predictions
        visualize_predictions(y_train, y_train_pred, y_val, y_val_pred, model_params['name'])

    # Convert the list of results to a DataFrame
    results_df_A = pd.DataFrame(results_list)

    return results_df_A

# Assuming freq, p, d, q, P, D, Q, m are defined elsewhere
# Call the function with your y_train, X_train, y_val, and X_val data
results_df_A = generate_sarimax_leaderboard(y_train, X_train, y_val, X_val, p, d, q, P, D, Q, m)

# Sort results by RMSE
results_df_A = results_df_A.sort_values(by=['RMSE_Val', 'RMSE_Train'])

results_df_A
```

##### Time Series Modeling using parameters selected via smart search

```python
def find_best_sarimax_parameters(train_data, exog_train, seasonal=True, m=selected_period):
    auto_arima_model = auto_arima(train_data, exogenous=exog_train, seasonal=seasonal, m=selected_period, freq=freq, suppress_warnings=True)
    best_order = auto_arima_model.order
    best_seasonal_order = auto_arima_model.seasonal_order

    return best_order, best_seasonal_order

def fit_sarimax_model(train_data, exog_train, order, seasonal_order):
    model = SARIMAX(train_data, exog=exog_train, order=order, seasonal_order=seasonal_order, freq=freq, enforce_stationarity=False)
    fitted_model = model.fit(disp=False, method='lbfgs', maxiter=100)
    return fitted_model

# Function to evaluate model performance
def evaluate_model(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    return mse, rmse, mae

# Function to visualize predictions
def visualize_predictions(y_train, pred_train, y_val, pred_val):
    plt.figure(figsize=(11, 2))
    plt.plot(y_train.index, y_train, label='Actual', color='blue')
    plt.plot(y_train.index, pred_train, label='Training Set Prediction', color='red', linestyle='--')
    plt.plot(y_val.index, y_val, label='Actual (Validation)', color='green')
    plt.plot(y_val.index, pred_val, label='Validation Set Prediction', color='orange', linestyle='--')
    plt.title('SARIMAX - Actual vs Predicted')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()
    plt.show()

def generate_sarimax_leaderboard(y_train, exog_train, y_val, exog_val):
    results_list = []

    # Find the best parameters
    best_order, best_seasonal_order = find_best_sarimax_parameters(y_train, exog_train)

    # Fit SARIMA model with the best parameters
    sarimax_model = fit_sarimax_model(y_train, exog_train, best_order, best_seasonal_order)
    
    # Generate predictions for training and validation sets
    y_train_pred = sarimax_model.predict(start=0, end=len(y_train) - 1, exog=exog_train)
    y_val_pred = sarimax_model.get_forecast(steps=len(y_val), exog=exog_val).predicted_mean

    # Evaluate model performance
    mse_train, rmse_train, mae_train = evaluate_model(y_train, y_train_pred)
    mse_val, rmse_val, mae_val = evaluate_model(y_val, y_val_pred)
    
    # Append results to list
    results_list.append({
        'Model': 'SARIMAX',
        'Params': f'Order={best_order}, Seasonal_Order={best_seasonal_order}',
        'AIC': sarimax_model.aic,
        'MSE_Train': mse_train,
        'RMSE_Train': rmse_train,
        'MAE_Train': mae_train,
        'MSE_Val': mse_val,
        'RMSE_Val': rmse_val,
        'MAE_Val': mae_val
    })
    
    # Visualize predictions
    visualize_predictions(y_train, y_train_pred, y_val, y_val_pred)

    # Create DataFrame from the list of dictionaries
    results_df = pd.DataFrame(results_list)

    return results_df


# Assuming y_train, X_train, y_val, and X_val are defined elsewhere
results_df_C = generate_sarimax_leaderboard(y_train, X_train, y_val, X_val)

# Display the results DataFrame
results_df_C
```
