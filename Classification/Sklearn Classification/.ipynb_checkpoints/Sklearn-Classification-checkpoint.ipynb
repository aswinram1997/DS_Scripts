```python
import pandas as pd
import numpy as np
import sklearn
from scipy.stats import randint, uniform

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn import metrics
from sklearn.metrics import accuracy_score, f1_score

import joblib
```

### Data Collection

```python
# read csv
df = pd.read_csv('Telco-Customer-Churn.csv')

# drop customer_id
df = df.drop(columns=['customerID'], axis = 1)

# label encode target
le = LabelEncoder()
df['Churn'] = le.fit_transform(df['Churn'])

df
```

```python
df = df.sample(n=500, random_state=42)

df
```

### Data Split

```python
df.columns
```

```python
X = df[['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod',
       'MonthlyCharges', 'TotalCharges']]

y = df['Churn']

# create numerical and categorical columns
numerical_columns = X.select_dtypes(include=['int64','float64']).columns.tolist()
categorical_columns = X.select_dtypes(include=['object']).columns.tolist()
```

```python
# Create train split
X_train, X_rest, y_train, y_rest = train_test_split(X, y, train_size=0.8, random_state=42, stratify=y)

# Create val and test split
X_val, X_test, y_val, y_test = train_test_split(X_rest, y_rest, train_size=0.5, random_state=42, stratify=y_rest)


print(f"Shape of X_train: {X_train.shape[0]/X.shape[0]}%")
print(f"Shape of X_val: {X_val.shape[0]/X.shape[0]}%")
print(f"Shape of X_test: {X_test.shape[0]/X.shape[0]}%")
```

### Data Preprocessing

```python
# handling numerical columns
# feature scaling
scaler = MinMaxScaler()
X_train_numerical = scaler.fit_transform(X_train[numerical_columns])
X_val_numerical = scaler.transform(X_val[numerical_columns])
X_test_numerical = scaler.fit_transform(X_test[numerical_columns])
```

```python
# handle categorical columns
# label encoding
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output = False)
X_train_categorical = encoder.fit_transform(X_train[categorical_columns])
X_val_categorical = encoder.transform(X_val[categorical_columns])
X_test_categorical = encoder.fit_transform(X_test[categorical_columns])
```

```python
# horizontally stack columns
X_train = np.hstack((X_train_numerical, X_train_categorical))
X_val = np.hstack((X_val_numerical, X_val_categorical))
X_test = np.hstack((X_test_numerical, X_test_categorical))
```

```python
# check count of target
df['Churn'].value_counts()
```

```python
# random oversampling
ros = RandomOverSampler(random_state=42)
X_oversampled, y_oversampled = ros.fit_resample(X_train, y_train)

# random undersampling
rus = RandomUnderSampler(random_state=42)
X_undersampled, y_undersampled = ros.fit_resample(X_train, y_train)

# random oversampling
sm = SMOTE(random_state=42)
X_smote, y_smote = sm.fit_resample(X_train, y_train)
```

### Model Training and Validation

```python
# Create algorithm objects as a list
algorithms = [LogisticRegression(), SVC(), RandomForestClassifier()]

# Define hyperparameter grid
param_grid_lr = {}
param_grid_svc = {'C': uniform(0.1, 1), 'kernel': ['linear', 'poly', 'rbf']}
param_grid_rf = {'n_estimators': randint(10, 100), 'max_depth': randint(10, 35), 'criterion': ['gini', 'entropy', 'log_loss']}
hyperparam_grid = [param_grid_lr, param_grid_svc, param_grid_rf]

# Random search objects
for algorithm, hyperparams in zip(algorithms, hyperparam_grid):
    # Create random search object
    rscv = RandomizedSearchCV(algorithm, hyperparams, random_state=0, n_iter=10, cv=5, scoring='accuracy')
    
    # Create a model with the random search object
    clf = rscv.fit(X_smote, y_smote)

    # Print the results 
    print(f'Best Parameters for {algorithm.__class__.__name__}: {clf.best_params_}')
    print(f'Best Score for {algorithm.__class__.__name__}: {round(clf.best_score_, 2)}')
    print('')
```

```python
# create best_model object
best_model = RandomForestClassifier(criterion = 'log_loss', max_depth = 14, n_estimators = 97)
best_model.fit(X_smote, y_smote)

# predict on train, val and test set
y_train_pred = best_model.predict(X_smote)
y_val_pred = best_model.predict(X_val)
y_test_pred = best_model.predict(X_test)

# get perfromance on train, val and test set
print(f'Performance on Train Set: {accuracy_score(y_smote, y_train_pred)}')
print(f'Performance on Val Set: {accuracy_score(y_val, y_val_pred)}')
print(f'Performance on Test Set: {accuracy_score(y_test, y_test_pred)}')
```

### Model Save

```python
# save preprocessing steps
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(encoder, 'encoder.pkl')

# save best model
joblib.dump(best_model, 'best_model.pkl')

print('Model Saved successfully!')
```
